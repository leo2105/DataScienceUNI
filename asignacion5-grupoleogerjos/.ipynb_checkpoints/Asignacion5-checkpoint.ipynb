### Visualización de datos en Python

Las visualizaciones son una herramienta fundamental para entender y compartir ideas sobre los datos. En python existen muchas  librerías para visualizar datos. Al tener tanta variedad de opciones, a veces se hace realmente difícil determinar cuando utilizar cada una de ellas.

### Algunos algoritmos del Machine learning


#### Aprendizaje Supervisado

Los algoritmos de aprendizaje supervisado hacen predicciones basadas en un conjunto de ejemplos. Por ejemplo, los precios históricos de las acciones pueden usarse para hacer estimaciones aventuradas de los precios futuros. Cada ejemplo usado para el entrenamiento se etiqueta con el valor de interés; en este caso, el precio de las acciones. Un algoritmo de aprendizaje supervisado busca patrones en esas etiquetas de valor. Puede usar cualquier información que pueda ser relevante, como el día de la semana, la temporada, datos financieros de la empresa, el tipo de sector o la presencia de eventos geopolíticos perjudiciales, y cada algoritmo busca tipos diferentes de patrones. Una vez que el algoritmo encuentra el mejor patrón posible, lo usa para hacer predicciones de datos de prueba sin etiquetar; en este caso, los precios futuros.


* Clasificación. Cuando los datos se usan para predecir una categoría, el aprendizaje supervisado también se denomina clasificación. Esto ocurre cuando se asigna una imagen, como una foto de un 'gato' o un 'perro'. Cuando hay solo dos opciones, se denomina clasificación de dos clases o binomial. Cuando hay más categorías, como cuando se predice el ganador del torneo de la NCAA, este problema se conoce como clasificación multiclase.

* Regresión. Cuando se predice un valor, como el precio de las acciones, el aprendizaje supervisado se denomina regresión.


#### Aprendizaje no  supervisado

En el aprendizaje no supervisado los puntos de datos no tienen etiquetas asociadas a ellos. En su lugar, el objetivo de un algoritmo de aprendizaje sin supervisar es organizar los datos de alguna manera o describir su estructura. Esto puede significar agruparlos en clústeres o buscar diferentes maneras de examinar datos complejos para que parezcan más simples o más organizados.

#### Aprendizaje de refuerzo

En el aprendizaje de refuerzo, el algoritmo elige una acción en respuesta a cada punto de datos. El algoritmo de aprendizaje también recibe una señal de recompensa un poco más adelante, que indica cómo de buena fue la decisión. Según esto, el algoritmo modifica su estrategia para lograr la mayor recompensa. El aprendizaje de refuerzo es común en robótica, donde el conjunto de lecturas del sensor en un punto en el tiempo es un punto de datos, y el algoritmo debe elegir la siguiente acción del robot. También resulta perfecto para las aplicaciones de Internet de las cosas.

4 . Presentar un reporte del siguiente algoritmo:

```python
import math
import heapq
import random

def ClusterGauss(centro, stdDev, count=50):
    # Devuelve 50 puntos donde cada puntos se representa como par ordenado (x, y).
    # Cada 'x' toma un valor en la distribucion gaussina donde tiene media centro[0 o 1] y desv. estandar stdDev.
    # El valor de 'y' se toma igual que al valor de 'x'. 
    return [(random.gauss(centro[0], stdDev),
             random.gauss(centro[1], stdDev)) for _ in range(count)]

def Haciendodatos_ficticios():
    # Retornamos 2 arreglos de size 50 cada uno.
    return ClusterGauss((-4,0), 1) + ClusterGauss((4,0), 1)

def DistanciaEuclidea(x,y):
    # x : Punto pasado por argumento
    # y : Punto de entrenamiento
    # Calcula la distancia euclidiana en una sola linea.
    return math.sqrt(sum([(a-b)**2 for (a,b) in zip(x,y)]))

def ClasificadorKNN(data, etiquetas, k, distancia):
    # Se crea una funcion que permite clasificar un punto x si se debe etiquetar como 1 o 2.
    def clasificar(x):
        # "Puntoscercanos" son los k puntos mas cercanos a x dentro del conjunto de puntos data, 
        # usando la distancia euclideana.
        Puntoscercanos = heapq.nsmallest(k, enumerate(data), key=lambda y: distancia(x, y[1]))
        # "Etiquetascercanas son las etiquetas de los puntos mas cercanos"
        Etiquetascercanas = [etiquetas[i] for (i, pt) in Puntoscercanos]
        # Se devuelve la etiqueta con mayor repeticion
        return max(set(Etiquetascercanas), key=Etiquetascercanas.count)
    return clasificar

if __name__ == "__main__":
    import sys
    
    k = sys.argv[1] if len(sys.argv) == 2 else 8
    Puntosentrenamiento = Haciendodatos_ficticios()
    Etiquetasentrenamiento = [1] * 50 + [2] * 50  
    
    # Hasta ahora se crearon puntos aleatorios y cada uno con su etiqueta.
    
    # f es el funcionq que clasifica el punto ingresado como argumento. Claramente se ve que es aprendizaje 
    # supervisado, debido a las etiquetas asignadas.
    f = ClasificadorKNN(Puntosentrenamiento, Etiquetasentrenamiento, k, DistanciaEuclidea)
    
    # Veremos a que etiqueta pertenecen estos puntos.
    print (f((-3,0)))
    print (f((3,0)))
    print (f((0,0)))
```

5 . Presentar un reporte del siguiente algoritmo.

```python
import numpy
from numpy import array, transpose, dot
from numpy.linalg import inv

# función que nos devuelve la media de una lista
avg = lambda L: 1.0 * sum(L)/len(L)

# regresión lineal de una variable de entrada, una variable de salida
def mejorEstimadorLineal(puntos):
    # 
    xAvg, yAvg = map(avg, zip(*puntos))
    aNum = 0
    aDenom = 0
    for (x,y) in puntos:
        aNum += (y - yAvg) * x
        aDenom += (x - xAvg) * x
        
    a = float(aNum) / aDenom
    b = yAvg - a * xAvg
    return (a, b), lambda x: a*x + b

# Este algoritmo usa la regresión contraida o ride regression 
# regresión lineal que puede recibir entrada de forma multivariable y nos resulta una variable de salida
def mejorEstimadorLinealMV(puntos):
    #X es matriz y w e y son coeficientes
    X = array([[1] + list(p[:-1]) for p in puntos]) 
    y = array([p[-1] for p in puntos])
    
    Xt = transpose(X)
    theInverse = inv(dot(Xt, X))
    w = dot(dot(theInverse, Xt), y)
    return w, lambda x: dot(w, x)


if __name__ == "__main__":
    
    import random
    
    # Generamos un conjunto de puntos aleatorios
    # Aquí los puntos se generan aleatoriamente cerca a la pendiente 0.5 e intercepto 7.0
    a = 0.5
    b = 7.0
    noise = lambda: random.random() * 0.4 - 0.2
    
    # El resultado de la función mejorEstimadorLineal será (pendiente, intercepto) y será cercano al
    # valor a y b declarado arriba lo que significa que hemos encontrado una mejor ecuación de línea con menos error.
    
    puntos = [(x, a*x + b + noise()) for x in [random.random() * 10 for _ in range(100)]]
    print(mejorEstimadorLineal(puntos)[0])
    
    # Use esta parte para plotear los puntos y la recta obtenida por el estimador lineal
    X = []
    Y = []
    for i in puntos:
        X.append(i[0])
        Y.append(i[1])
        
    l = np.arange(0,10)
    line = []
    
    a = mejorEstimadorLineal(puntos)[0][0]
    b = mejorEstimadorLineal(puntos)[0][1]
    for i in range(len(l)):
        line.append(a*l[i]+b)

    plt.scatter(X,Y)
    plt.plot(line,'r')
    plt.show()

    print(mejorEstimadorLinealMV(puntos)[0])
    
    l = np.arange(0,10)
    line = []
    b = mejorEstimadorLinealMV(puntos)[0][0]
    a = mejorEstimadorLinealMV(puntos)[0][1]
    for i in range(len(l)):
        line.append(a*l[i]+b)
    
    plt.scatter(X,Y)
    plt.plot(line,'g')
    plt.show()
    
    # En este ejemplo usamos 6 dimensiones para nuestro regresor lineal
    
    verdaderoW = array([-3,1,2,3,4,5])
    bias, term_lineal = verdaderoW[0], verdaderoW[1:]
    
    puntos = [tuple(v) + (dot(term_lineal, v) + bias + noise(),) for v in [numpy.random.random(5) for _ in range(100)]]
    
    print(mejorEstimadorLinealMV(puntos)[0])
```

6 .Reduzca la dimensión de un conjunto de datos, traduciendo cada punto de datos en una representación que capture las características "más importantes". Explica tu respuesta.

```python
# Tu codigo

#Para este ejercicio, utilizaremos el dataset relacionado con la prevalencia de Tuberculosis Infecciosa:

import pandas as pd
    
existing_df = pd.read_csv('Datos/indicator_estimated.csv', 
    index_col = 0, 
    thousands  = ',')
existing_df.index.names = ['country']
existing_df.columns.names = ['year']
existing_df.head()
```

### Reducción dimensional con PCA

En nuestro conjunto de datos, cada muestra es un país definido por 18 variables diferentes, cada una correspondiendo a casos de tuberculosis por 100K (existentes, nuevos, muertes) para un año dado de 1990 a 2007. Mediante el uso de [PCA](https://en.wikipedia.org/wiki/Dimensionality_reduction#Principal_component_analysis_.28PCA.29) (Principal Component Analisis) seremos capaces de reducir estas 18 variables sólo a las dos que mejor captura esa información.

Para usaremos la librería de sklearn que ya viene con una implementacion PCA. Esta implementación utiliza la implementación scipy.linalg de la descomposición de valores singulares. Sólo funciona para matrices densas y no es escalable a datos dimensionales grandes. Para datos dimensionales grandes debemos considerar algo como las características de reducción de dimensionalidad de Spark.

```python
from sklearn.decomposition import PCA
    
pca = PCA(n_components=2)
pca.fit(existing_df)
```

Esto nos da un objeto que podemos usar para transformar nuestros datos llamando transform.

```python
existing_2d = pca.transform(existing_df)
```

Lo colocamos en un nuevo dataframe

```python
existing_df_2d = pd.DataFrame(existing_2d)
existing_df_2d.index = existing_df.index
existing_df_2d.columns = ['PC1','PC2']
existing_df_2d.head()
```

Ahora vamos a graficar la versión reducida de dimensiones de nuestro conjunto de datos.

```python
ax = existing_df_2d.plot(kind='scatter', x='PC2', y='PC1', figsize=(16,8))
    
for i, country in enumerate(existing_df.index):
    ax.annotate(
        country, 
        (existing_df_2d.iloc[i].PC2, existing_df_2d.iloc[i].PC1)
    )
```
