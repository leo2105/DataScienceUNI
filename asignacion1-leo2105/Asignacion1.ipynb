9 . En que consiste el protocolo interador (The Iterator Protocol) de Python.

Usar el `iterator protocol` es la unica manera universal para recorrer un iterable(cualquier objeto que se puede 
recorrer con un loop **for**) en python.

El `iterator protocol` nos dice que podemos recorrer un iterable mediante un iterator con una funcion **next** del objeto
iterator y nos retorn una excepcion `StopIteration` cuando hemos terminado de recorrer el iterable.
